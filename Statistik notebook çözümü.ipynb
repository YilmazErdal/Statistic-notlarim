## Mean - Median - Mode - Skewness

```python
import numpy as np
from scipy import stats
```

```python
age = [20, 22, 25, 25, 27, 27, 27, 29, 30, 31, 121]
```

```python
np.mean(age)  #  age datasının ortalamasını alıyoruz
```

```python
np.median(age)  # age veri setinin medianını hesapladık
```

```python
stats.mode(age) # age veri setinin mod unu hesaplar.mod işlemini stats kütüphanesinden mod fonksiyonunu çağırarak yapıyoruz
```

```python
stats.mode(age)[0]   # direkt mod  sayısını getiriyor. Mod en çok tekrar eden eleman demek.
```

```python
age2 = [20, 22, 25, 25, 27, 27, 27, 29, 30, 31, 121]
```

```python
np.max(age2) - np.min(age2)   # range  bulduk.
```

```python
# Variance
np.var(age2)   # age2 nin varyansını hesapladık.
```

```python
# IQR

from scipy.stats import iqr    # iqr ı scipy.stats kütüphanesinden çektik
```

```python
iqr(age2)
```

```python
# Q1 ve Q3 hesabı

q1 = np.percentile(age2,25)  # 25 Q1 i verir. 50 Q2 yi verir. 75 Q3 ü verir.
q1
```

```python
q3 = q1 = np.percentile(age2,75)
q3
```

```python
iqr = (q3 - q1)
iqr
```

##  Skewness and Kurtosis

```python
# kurtosis

stats.kurtosis(age2)   # Kurtosis  basıslık değerini veriyor.
```

```python
# skewness
stats.skew(age2)
```

## Population and Sample

```python
np.random.seed(101)  #seed ile heipimiz için aynı random degreler uretıyoruz.

population = np.random.randint(0,100, size = 1000)
```

```python
population
```

```python
# sampling
np.random.seed(51)  #seed ile heipimiz için aynı random degreler uretılıyor

sample = np.random.choice(population,100)
```

```python
sample
```

```python
population.mean()
```

```python
sample.mean()
```

```python
sample_means = []    #  liste oluşturduk.
for i in range(100):
  sample = np.random.choice(population, 100)
  sample_means.append(sample.mean())
sample_means
```

```python
np.mean(sample_means)
```
